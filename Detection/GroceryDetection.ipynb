```python
import os

def print_directory_tree(dir_path, indent=0):
    indent_str = '    ' * indent
    print(f"{indent_str}{os.path.basename(dir_path)}/")
    try:
        with os.scandir(dir_path) as entries:
            subdirs = []
            file_count = 0

            for entry in entries:
                if entry.is_dir():
                    subdirs.append(entry.name)
                else:
                    file_count += 1
            for subdir in subdirs:
                print_directory_tree(os.path.join(dir_path, subdir), indent + 1)
            if not subdirs:
                print(f"{indent_str}    [Files: {file_count}]")
    except PermissionError:
        print(f"{indent_str}    [Permission Denied]")

directory_path = "/content/Grocery_Items-45"
print_directory_tree(directory_path)
```

```python
import os
import glob

def analyze_label_file(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()

    box_count = 0
    seg_count = 0
    for line in lines:
        parts = line.strip().split()
        if len(parts) == 5:  # class_id, x, y, width, height
            box_count += 1
        elif len(parts) > 5:  # class_id, x1, y1, x2, y2, ...
            seg_count += 1

    return box_count, seg_count

def analyze_dataset(dataset_path):
    label_files = glob.glob(os.path.join(dataset_path, '**', '*.txt'), recursive=True)

    total_box_count = 0
    total_seg_count = 0
    mismatch_files = []

    for file in label_files:
        box_count, seg_count = analyze_label_file(file)
        total_box_count += box_count
        total_seg_count += seg_count

        if box_count != seg_count:
            mismatch_files.append(file)

    return total_box_count, total_seg_count, mismatch_files

# Analyze both train and validation sets
train_path = '/content/Grocery_Items-45/train/labels'
val_path = '/content/Grocery_Items-45/valid/labels'

for path in [train_path, val_path]:
    boxes, segs, mismatches = analyze_dataset(path)
    print(f"Analysis for {path}:")
    print(f"Total bounding boxes: {boxes}")
    print(f"Total segmentations: {segs}")
    print(f"Number of files with mismatches: {len(mismatches)}")
    print(f"Sample mismatched files: {mismatches[:5]}\n")
```

```python
import os
import glob

def seg_to_box(seg_points):
    x_coords = seg_points[::2]
    y_coords = seg_points[1::2]
    x_min, y_min = min(x_coords), min(y_coords)
    x_max, y_max = max(x_coords), max(y_coords)
    return [(x_min + x_max) / 2, (y_min + y_max) / 2, x_max - x_min, y_max - y_min]

def convert_file(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()

    new_lines = []
    for line in lines:
        parts = line.strip().split()
        if len(parts) > 5:
            class_id = parts[0]
            box = seg_to_box(list(map(float, parts[1:])))
            new_lines.append(f"{class_id} {' '.join(map(str, box))}\n")
        else:  # It's already a box
            new_lines.append(line)

    with open(file_path, 'w') as f:
        f.writelines(new_lines)

def convert_dataset(dataset_path):
    label_files = glob.glob(os.path.join(dataset_path, '**', '*.txt'), recursive=True)
    for file in label_files:
        convert_file(file)

# Convert both train and validation sets
train_path = '/content/Grocery_Items-45/train/labels'
val_path = '/content/Grocery_Items-45/valid/labels'

for path in [train_path, val_path]:
    convert_dataset(path)
    print(f"Converted {path}")
```
